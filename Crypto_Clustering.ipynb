```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_market_data)
```

```python
# Create a DataFrame with the scaled data
# Copy the crypto names from the original DataFrame
# Set the coin_id column as index
df_scaled = pd.DataFrame(scaled_data, columns=df_market_data.columns, index=df_market_data.index)
# Display the scaled DataFrame
df_scaled.head()
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k_values = range(1, 12)
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for k in k_values:
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(df_scaled)
    inertia.append(model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k_values, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame({"k": k_values, "inertia": inertia})

# Display the Elbow DataFrame
df_elbow.head()
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(
    x="k",
    y="inertia",
    title="Elbow Method for Optimal k",
    xlabel="Number of Clusters (k)",
    ylabel="Inertia",
    width=800,
    height=400
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** From the chart, the Elbow curve shows a significant decrease in inertia up to k = 4, after which the rate of decrease becomes much smaller.
Therefore, the best value for k is 4.

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-means model using the best value for k
best_k = 4  # Use the optimal value determined from the Elbow curve
model = KMeans(n_clusters=best_k, random_state=42)

# Fit the K-means model using the scaled DataFrame
model.fit(df_scaled)

# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
df_scaled['Cluster'] = model.predict(df_scaled)

# Display the first few rows of the DataFrame with the cluster labels
df_scaled.head()
```

```python
# Fit the K-Means model using the scaled DataFrame
model.fit(df_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
df_scaled['Cluster'] = model.predict(df_scaled)


# Print the resulting array of cluster values.
print(df_scaled['Cluster'].unique())
```

```python
# Create a copy of the scaled DataFrame
df_clustered = df_scaled.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
df_clustered['Cluster'] = model.labels_
# Display the copy of the scaled DataFrame
df_clustered.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
df_scaled.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="Cluster",
    hover_cols=["coin_id"],
    title="Cryptocurrency Clusters (Original DataFrame)"
)
```

### Observations:
- Cluster 0 contains cryptocurrencies with relatively stable performance (low 24-hour and 7-day price changes).
- Cluster 1 appears to contain volatile cryptocurrencies with significant price changes.
- Cluster 2 and 3 represent intermediate behaviors, with Cluster 3 having slightly higher 24-hour price changes.

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set n_components=3
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.
pca_data = pca.fit_transform(df_scaled.drop(columns=["Cluster"]))

# Convert PCA-transformed data into a DataFrame for easier handling
df_pca = pd.DataFrame(
    pca_data,
    columns=["PC1", "PC2", "PC3"],
    index=df_scaled.index
)
# Display the PCA DataFrame
df_pca.head()
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
explained_variance = pca.explained_variance_ratio_.sum()

# Print the total explained variance as a percentage
print(f"Total Explained Variance: {explained_variance:.2%}")
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total explained variance of the three principal components is 89.50%

```python
# Create a new DataFrame with the PCA data.
df_pca = pd.DataFrame(
    pca_data,
    columns=["PC1", "PC2", "PC3"],  # Name the principal components
    index=df_scaled.index           # Retain the original index (coin_id)
)

# Copy the crypto names from the original scaled DataFrame
# (This step is already implicitly done by using the index from df_scaled).

# Set the coin_id column as index
# (This is already handled above by setting index=df_scaled.index).

# Display the scaled PCA DataFrame
df_pca.head()

```

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
k_values = range(1, 12)
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_pca)
    inertia.append(kmeans.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k_values, "inertia": inertia}
# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(
    x="k",
    y="inertia",
    title="Elbow Method for Optimal k (PCA Data)",
    xlabel="Number of Clusters (k)",
    ylabel="Inertia",
    width=800,
    height=400
)
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:**The best value for k is 4, as the elbow - where the inertia drops significantly before flattening out - occurs at k=4.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No, the best value for k using the PCA data is the same as the value found using the original data (k=4).

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialize the K-Means model using the best value for k
best_k = 4  # Use the optimal value determined from the PCA Elbow curve
kmeans_pca = KMeans(n_clusters=best_k, random_state=42)
```

```python
# Fit the K-Means model using the PCA data
kmeans_pca.fit(df_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled PCA DataFrame
df_pca['Cluster'] = kmeans_pca.predict(df_pca)

# Print the resulting array of cluster values.
df_pca.head()
```

```python
# Create a copy of the scaled PCA DataFrame
df_pca_copy = df_pca.copy()

# Remove the Cluster column if it already exists in the DataFrame
if 'Cluster' in df_pca_copy.columns:
    df_pca_copy = df_pca_copy.drop(columns=["Cluster"])

# Add a new column to the copy of the PCA DataFrame with the predicted clusters
df_pca_copy['Cluster'] = kmeans_pca.predict(df_pca_copy)

# Display the copy of the scaled PCA DataFrame
df_pca_copy.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="PC1"` and `y="PC2"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
df_pca_copy.hvplot.scatter(
    x="PC1",
    y="PC2",
    by="Cluster",
    hover_cols=["coin_id"],
    title="Cryptocurrency Clusters (PCA DataFrame)"
)
```

#### Answer the following question: 

  * **Question:** After visually analyzing the cluster analysis results, what is the impact of using fewer features to cluster the data using K-Means?

  * **Answer:**  Using fewer features (via PCA) for clustering simplifies the data structure and reduces noise from less significant features. This optimization results in:

Improved Cluster Visualization:

The PCA-transformed data provides a cleaner separation of clusters in the 2D plot (PC1 vs. PC2), making it easier to interpret.
Reduced Dimensionality:

With PCA, the clustering relies on a smaller number of components while retaining most of the variance, thus reducing computational complexity.
Slightly Altered Cluster Assignments:

While most cryptocurrencies remain in the same clusters, minor differences may arise due to reduced feature space.
Trade-off Between Simplicity and Precision:

PCA-based clustering offers a simplified, more interpretable model but might lose some granularity compared to clustering on the full feature set.
